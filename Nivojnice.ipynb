#Iskanje nivojnic
Na sliki so predstavljene številčne vrednosti. Obris je tam, kjer je vrednost enaka 0. Poišči obris.

```python
x = linspace(0,1,10)
y = x
X, Y = meshgrid(x,y)
fig, ax = subplots()
ax.plot(X, Y, 'b.')
axis("off")
Z = 10*(rand(10,10)-0.5)
# The key option here is `bbox`. I'm just going a bit crazy with it.
for i in range(10):
    for j in range(10):
        ax.annotate("%1.1f" % Z[i,j], xy=(x[i], y[j]), xytext=(5,5), 
            textcoords='offset points', ha='center', va='bottom')
```

```python
contour(Z.T,levels=[0])
```

## Riba
Sestavimo sedaj funkcijo, ki bo implicitno določala preprosto podobo ribe.

```python
f = lambda x,y: exp(-(3*(x-0.75)**2+2*(y-0.5)**2))-0.9 + exp(-9*x**2-(y-0.5)**2)
Z = 10*f(X,Y)
cp = contour(Z)
cp.clabel(fmt='%1.1f',inline=1,fontsize=10)
```

```python
fig, ax = subplots()
ax.plot(X, Y, 'b.')
axis("off")
# The key option here is `bbox`. I'm just going a bit crazy with it.
for i in range(10):
    for j in range(10):
        ax.annotate("%1.1f" % Z[i,j], xy=(x[i], y[j]), xytext=(5,5), 
            textcoords='offset points', ha='center', va='bottom')
```
