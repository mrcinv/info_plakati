#Iskanje nivojnic
Na sliki so predstavljene številčne vrednosti. Obris je tam, kjer je vrednost enaka 0. Poišči obris.

```python
%pylab inline
n = 10
x = linspace(0,1,n)
y = x
X, Y = meshgrid(x,y)
fig, ax = subplots()
ax.plot(X, Y, 'b.')
axis("off")
Z = 10*(rand(n,n)-0.5)
# stevilke poleg pikic
for i in range(n):
    for j in range(n):
        ax.annotate("%1.1f" % Z[i,j], xy=(x[i], y[j]), xytext=(5,5), 
            textcoords='offset points', ha='center', va='bottom')
```

```python
contour(Z.T,levels=[0])
```

## Riba
Sestavimo sedaj funkcijo, ki bo implicitno določala preprosto podobo ribe.

```python
f = lambda x,y: exp(-(3*(x-0.75)**2+2*(y-0.5)**2))-0.9 + exp(-9*x**2-(y-0.5)**2)
Z = 10*f(X,Y)
cp = contour(Z)
cp.clabel(fmt='%1.1f',inline=1,fontsize=10)
```

# Radialne bazne funkcije
Dani obris (podan s poligonom) lahko spremenimo v implicitno obliko, tako da sestavimo interpolacijo z *radialnimi baznimi funkcijami* oblike

$$\newcommand{\x}{\mathbf{x}}R_{\x_0}(\x)=\varphi(||\x_0 -\x||),$$

kjer je $\varphi$ funkcija s posebnimi lastnostmi
  
  1. $\varphi'(0)=0$ in $\varphi'(r)\le 0$ za $r>0$
  2. $\lim_{r\to\infty}\varphi(r)=0$

Interpolacijsko funkcijo sestavimo kot linearno kombinacijo

$$f(\x) = \sum_{\x_i}\alpha_i R_{x_i}(\x)=\sum_{\x_i}\alpha_i \varphi(||\x_i-\x||)$$
