#Iskanje najtoplejšega kotička v hiši
Robotkove baterije v mrazu delujejo slabše, zato išče vir toplote, da bi se ogrel. Na srečo ima infrardečo kamero, ki pa se je zataknila in jo lahko usmeri le v tla. Poleg tega deluje potrebuje veliko časa, da posname sliko. Pomagaj mu poiskati peč v garaži.

```python
# temperaturna funkcija, verjetno bi lahko bila bolj enostavna
temp = lambda X,Y: 5*exp(-((X+5)**2/10+(Y-2)**2/10))+10*exp(-sqrt((X-2)**2/4+(Y+1)**2/3))-2/(X+5)+2/(Y-4) +\
    40*exp(-sqrt(5*(X-2)**2+10*(Y+1)**2))
```

```python
%pylab inline
from mpl_toolkits.mplot3d import Axes3D
x = arange(-4,4,0.02)
y = arange(-3,3,0.02)
X,Y = meshgrid(x,y)
Z = temp(X,Y)
fig = figure()
#ax = fig.gca(projection='3d')
#ax.plot_surface(X,Y,Z)
cp = contour(X,Y,Z,levels=arange(-2,30))
cp.clabel(arange(-2,7),fmt='%1.0f',inline=1,fontsize=10)
im = imshow(Z, interpolation='bicubic',norm=None,vmin=-3,vmax=9,origin='lower',extent=(-4,4,-3,3))
colorbar(orientation='horizontal', shrink=0.8)
```

```python
#lokalna predstavitev temperaturne funkcije
def plot_temp(x0,y0):
    "Plots a contour plot in a small square around the center (x0,y0)"
    delta = 0.1
    x = arange(x0-delta,x0+delta,0.05)
    y = arange(y0-delta,y0+delta,0.05)
    X,Y = meshgrid(x,y)
    Z = temp(X,Y)
    extent = (x0-delta,x0+delta,y0-delta,y0+delta)
    #cp = contour(Z,levels=arange(-2,7,0.5),extent=extent)
    #cp.clabel(arange(-2,7),fmt='%1.0f',inline=1,fontsize=10)
    im = imshow(Z, interpolation='bicubic',origin='lower',extent=extent)
    axis('off')
    colorbar(orientation='horizontal', shrink=0.8)
    return gca()
#colorbar(orientation='horizontal', shrink=0.8)
#plot_temp(0,0)
#plot_temp(-1,1)
plot_temp(-4,-1)
```

```python
def grad(Z):
    """Numerical gradient of a matrix"""
    zy = mean(Z[1:,:]-Z[:-1,:])
    zx = mean(Z[:,1:]-Z[:,:-1])
    return array([zx,zy])
# Gradientna metoda
def korak(x0,y0):
    """En korak gradientne metode. Gradient se določi kot povprečni numerični gradient 
    v majhnem kvadratku okrog (x0,y0)"""
    delta = 0.1
    x = arange(x0-delta,x0+delta,0.05)
    y = arange(y0-delta,y0+delta,0.05)
    X,Y = meshgrid(x,y)
    Z = temp(X,Y)
    #contour(x,y,Z)
    #axis('equal')
    df = grad(Z)
    #plot([x0,x0 + df[0]],[y0,y0 + df[1]],'-o')
    #print df
    df = df/norm(df)
    return (x0+df[0],y0+df[1])
korak(-4,1)
```

#Iskanje najtoplejšega kotička v hiši
Robotkove baterije v mrazu delujejo slabše, zato išče vir toplote, da bi se ogrel. Na srečo ima infrardečo kamero, ki pa se je zataknila in jo lahko usmeri le v tla. Poleg tega kamera potrebuje veliko časa, da posname sliko.

Robotek posname prvo sliko v točki s koordinatami $(0,3)$. Vsakič, ko posname sliko, naredi en korak dolžine $1$ v smeri višje temperature. Po 5 korakih dobi naslednje zaporedje slik. Lahko poveš koordinate točke, kjer je vir toplote? 

```python
x0,y0 = (0,3)
x = [x0]
y = [y0]
figure(1)
plot_temp(x0,y0)
for i in range(5):
    x0,y0 = korak(x0,y0)
    x.append(x0)
    y.append(y0)
    figure(i+2)
    plot_temp(x0,y0)
```
